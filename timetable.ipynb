```python
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle, Paragraph
from reportlab.lib.pagesizes import A4, landscape
from reportlab.lib import colors
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle

file_path = "./Personal_Timetable_ECE_6C.pdf"

doc = SimpleDocTemplate(
    file_path,
    pagesize=landscape(A4),
    rightMargin=20,
    leftMargin=20,
    topMargin=20,
    bottomMargin=20
)

styles = getSampleStyleSheet()

title = Paragraph(
    "<b>Personal Timetable – ECE Section 6C</b>",
    styles["Title"]
)

cell_style = ParagraphStyle(
    "cell",
    parent=styles["Normal"],
    fontSize=9,
    leading=11,
    alignment=1  # center
)

def cell(text):
    return Paragraph(text, cell_style) if text else ""

data = [
    ["Day", "08–09", "09–10", "10–11", "11–12", "12–01",
     "01–02", "02–03", "03–04", "04–05", "05–06"],

    ["Monday", "", "", "",
     cell("VLSI<br/>(TW2GF2)"),
     "", "",
     cell("Embedded<br/>(TW2GF2)"),
     "", "",
     cell("Soft Computing<br/>(TW2GF1)")
    ],

    ["Tuesday",
     cell("TDD<br/>(TW2GF1)"),
     cell("TDD<br/>(TW2GF1)"),
     "",
     cell("VLSI<br/>(TW2GF2)"),
     cell("VLSI (VLSI Lab)"),
     cell("VLSI (VLSI Lab)"),
     cell("Embedded<br/>(TW2GF2)"),
     "", "", ""
    ],

    ["Wednesday",
     cell("Machine Learning<br/>(TW2GF1)"),
     cell("TDD<br/>(TW2GF1)"),
     "",
     "",
     "",
     "",
     cell("Soft Computing<br/>(TW2TF2)"),
     cell("Soft Computing<br/>(TW2GF1)"),
     "", ""
    ],

    ["Thursday",
     cell("TDD<br/>(TW2GF1)"),
     cell("TDD<br/>(TW2GF1)"),
     "",
     "",
     "", "",
     cell("Soft Computing<br/>(TW2TF2)"),
     cell("Soft Computing<br/>(TW2GF1)"),
     cell("ES Lab<br/>(MP LAB)"),
     cell("ES Lab<br/>(MP LAB)")
    ],

    ["Friday",
     cell("Machine Learning<br/>(TW2GF1)"),
     cell("TDD<br/>(TW2TF2)"),
     "",
     "",
     cell("Embedded<br/>(TW2GF2)"),
     cell("VLSI<br/>(TW2GF2)"),
     cell("Soft Computing<br/>(TW2TF2)"),
     cell("Soft Computing<br/>(TW2GF1)"),
     "",
     cell("Machine Learning<br/>(TW2TF2)")
    ],
]

col_widths = [75] + [70] * 10

table = Table(data, colWidths=col_widths, repeatRows=1)

table.setStyle(TableStyle([
    ("GRID", (0, 0), (-1, -1), 1, colors.black),
    ("BACKGROUND", (0, 0), (-1, 0), colors.lightgrey),
    ("ALIGN", (1, 1), (-1, -1), "CENTER"),
    ("VALIGN", (0, 0), (-1, -1), "MIDDLE"),
    ("FONTNAME", (0, 0), (-1, 0), "Helvetica-Bold"),
    ("BOTTOMPADDING", (0, 0), (-1, -1), 6),
    ("TOPPADDING", (0, 0), (-1, -1), 6),
]))

doc.build([title, table])

file_path
```
