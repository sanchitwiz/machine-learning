```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()  
```

```python
## missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[123:, "Region"] = 1
```

```python
dataset
```

```python
## i want to fill the region as int not float value
dataset['Region'] = dataset['Region'].astype(int)
```

```python
dataset.isnull().sum()
```

```python
## remove the null values
dataset = dataset.dropna().reset_index(drop=True)
dataset.info()
```

```python
dataset.iloc[[122]]
```

```python
dataset.drop(index=122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(index=122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
## fix spaces in column names
dataset.columns
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
## change un-neccessary column datatypes from float to int
dataset.info()
```

```python
dataset[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']]= dataset[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
## changing others to float datatype
objects = [features for features in dataset.columns if dataset[features].dtype == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        dataset[i] = dataset[i].astype(float)

dataset.info()
```

```python
## let's save the clean dataset
dataset.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

```python
## Exploratory Data Analysis
dataset_copy = dataset
```

```python
## drop date related columns for EDA
dataset_copy = dataset_copy.drop(columns=['day', 'month', 'year'], axis=1)
```

```python
dataset_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
dataset_copy['Classes'] = np.where(
    dataset_copy['Classes'].str.contains('not fire', case=False, na=False),
    0,
    1
)
```

```python
dataset_copy.head()
```

```python
dataset_copy['Classes'].value_counts()
```

```python
## Plot density of all features
dataset_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
## Percentage for pie chart
percentage = dataset_copy['Classes'].value_counts(normalize=True)*100
```

```python
## plotting pie chart
class_labels = ['Fire', 'Not Fire']
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=class_labels, autopct='%1.1f%%', startangle=140, colors=['orange', 'green'])
plt.title('Distribution of Fire vs Not Fire Cases')
```

```python
## Correlation
corr_matrix = dataset_copy.corr()
# print(corr_matrix)
dataset_copy.corr()
```

```python
sns.heatmap(corr_matrix)
```

```python
## Box Plots
sns.boxplot(dataset_copy['FWI'])
```

```python
dataset_temp = dataset.loc[dataset['Region'] == 0]

plt.subplots(figsize=(12, 6))
sns.set_style('whitegrid')

sns.countplot(
    x='month',
    hue='Classes',
    data=dataset_temp,
    palette='Set1'
)

plt.title("Monthly Fire Analysis")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```

```python
dataset_temp = dataset.loc[dataset['Region'] == 1]

plt.subplots(figsize=(12, 6))
sns.set_style('whitegrid')

sns.countplot(
    x='month',
    hue='Classes',
    data=dataset_temp,
    palette='Set1'
)

plt.title("Monthly Fire Analysis")
plt.xlabel("Month")
plt.ylabel("Count")
plt.show()
```
