## Hierarical Clustring Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
## import isis datasets
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(data=iris.data)
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
## Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_scaled_pca = pca.fit_transform(X_scaled)
```

```python
X_scaled_pca
```

```python
plt.scatter(X_scaled_pca[:,0], X_scaled_pca[:,1], c=iris.target)
```

```python
## Agglomerative Clustring
## to construct a dendogram

import scipy.cluster.hierarchy as sc

#plot the dendogram
plt.figure(figsize=(20, 7))
plt.title("Dendrograms")

#create dendogram
sc.dendrogram(sc.linkage(X_scaled_pca, method="ward"))
plt.xlabel("Sample Index")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
cluster.fit(X_scaled_pca)
```

```python
cluster.labels_
```

```python
plt.scatter(X_scaled_pca[:,0], X_scaled_pca[:,1], c=cluster.labels_)
```
