```python
import pandas as pd
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'New York'],
    'cost': [100, 80, 90, 70, 60, 50]
})
df
```

### On the basis of Mean

```python
mean_price = df.groupby('city')['cost'].mean().to_dict()
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
df
```

```python
df[['cost', 'city_encoded']]
```
