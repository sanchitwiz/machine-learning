## ðŸ”– Label Encoding

Label encoding involves assigning a unique **numerical label** to each category in a categorical variable.

### Example  
If we have a feature called **`color`**, label encoding might represent it as:

| Color | Encoded Value |
|-------|----------------|
| Red   | 1 |
| Blue  | 2 |
| Green | 3 |

> ðŸ”Ž **Note:**  
> - Label encoding is useful for **ordinal** data.  
> - For non-ordinal categories, it may introduce unintended ordering. In that case, prefer **One-Hot Encoding**.

```python
import pandas as pd
```

```python
## create a simple dataframe
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red', 'red', 'green']
})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

```python
lbl_encoder.transform([['green']])
```

# Odinal Encoding
#### Assigning Ranks
##### 1. High School: 1 `\n` 2. College: 2


```python
from sklearn.preprocessing import OrdinalEncoder
ord_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
df = pd.DataFrame({
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})

df.head()
```

```python
encoded_df = ord_encoder.fit_transform(df[['size']])    
encoded_df
```

```python
ord_encoder.transform([['large']])
```
