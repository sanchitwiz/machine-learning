## Handling Imbalance Dataset
#### Up Sampling
#### Down Sampling

```python
import pandas as pd
import numpy as np
```

```python
## create datapoints
np.random.seed(123)

## create a dataframe with 2 classes
n_samples = 1000
class_0_ration = 0.9
n_class_0 = int(n_samples * class_0_ration)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0, n_class_1
```

```python
## create my dataset with imbalance datasets
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    'feature_2': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'feature_2': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
class_0, class_1
```

```python
df = pd.concat([class_0, class_1]).reset_index(drop=True)
print(df)
```

```python
df['target'].value_counts()
```

### Up Samapling

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]

print(df_minority, df_majority)
```

```python
from sklearn.utils import resample
df_minority_upsampled = resample(df_minority, replace=True, n_samples = len(df_majority), random_state=42 )
print(df_minority.shape)
print(df_minority_upsampled.shape)
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsampled]).reset_index(drop=True)
print(df_upsampled)
```

## Down Sampling

```python
## create datapoints
np.random.seed(123)

## create a dataframe with 2 classes
n_samples = 1000
class_0_ration = 0.9
n_class_0 = int(n_samples * class_0_ration)
n_class_1 = n_samples - n_class_0


## create my dataset with imbalance datasets
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    'feature_2': np.random.normal(loc=0.0, scale=1.0, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'feature_2': np.random.normal(loc=2.0, scale=1.0, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

print(df['target'].value_counts())
```

```python
df_minority = df[df['target'] == 1]
df_majority = df[df['target'] == 0]

print(df_minority, df_majority)
```

```python
from sklearn.utils import resample
df_majority_downsampled = resample(df_majority, replace=False, n_samples = len(df_minority), random_state=42 )
print(df_majority.shape)
print(df_majority_downsampled.shape)
```

```python
df_sampled = pd.concat([df_majority_downsampled, df_minority]).reset_index(drop=True)
print(df_sampled)
```
