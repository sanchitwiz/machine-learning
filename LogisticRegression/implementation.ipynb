# Logistic Regression Impletation
Binary Classificaiton Problem

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
## create the dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
independent = pd.DataFrame(X)
dependent = pd.Series(y)
```

```python
independent.shape, dependent.shape
dependent.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.30, random_state=42)
```

```python
## Model Training
from sklearn.linear_model import LogisticRegression
logictic = LogisticRegression()
```

```python
logictic.fit(X_train, y_train)
```

```python
y_predict = logictic.predict(X_test)
print(y_predict)
```

```python
logictic.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score = accuracy_score(y_test, y_predict)
print("Accuracy:", score)
cm = confusion_matrix(y_test, y_predict)
print("Confusion Matrix:\n", cm)
cr = classification_report(y_test, y_predict)
print("Classification Report:\n", cr)
```

# Hyperparamteter tuning and Cross Validation 

```python
model = LogisticRegression()
penalty = ['l1','l2', 'elasticnet']
c_value = [0.01, 0.1, 1, 10, 100]
solver = ['liblinear', 'saga', 'newton-cg', 'lbfgs', 'sag']
```

```python
params = dict(penalty=penalty, C=c_value, solver=solver)
```

```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold()
grid = GridSearchCV(estimator=model, param_grid=params, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
print(grid)
## this will give you the best parameters that will fit in your model on the basis of accuracy
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score = accuracy_score(y_pred, y_test)
print("Accuracy:", score)
cm = confusion_matrix(y_pred, y_test)
print("Confusion Matrix:\n", cm)
cr = classification_report(y_pred, y_test)
print("Classification Report:\n", cr)
```

# Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
model = LogisticRegression()
random_cv = RandomizedSearchCV(estimator=model, param_distributions=params, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
random_cv.fit(X_train, y_train)
```

```python
random_cv.best_score_
```

```python
random_cv.best_params_
```

```python
y_pred = random_cv.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("Accuracy:", score)
cm = confusion_matrix(y_pred, y_test)
print("Confusion Matrix:\n", cm)
cr = classification_report(y_pred, y_test)
print("Classification Report:\n", cr)
```
